```python
items=['TV','냉장고','전자렌지','컴퓨터','선풍기','선풍기','믹서','믹서']
```

```python
# 레이블 인코딩
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
encoder.fit(items) # 내부적으로 번호에 데이터를 적용함 # 사용자에게 결과값을 주진 않는다.

labels = encoder.transform(items) # 인코딩 수행 labels 에서는 0,1,2,3 으로 들어가 있다.
```

```python
labels
```

```python
print("인코딩 변환 값 : {}".format(labels))
print("인코딩 클래스 : {}".format(encoder.classes_)) # class 값 인출
```

```python
# 디코딩 : 컴퓨터가 알아들을 수 있는 정보에서 사람이 알아 볼 수 있는 정보로 변환
# 인코딩의 반대
encoder.inverse_transform([1,5,4,2,3,2,1,2,3,4,0])
```

```python
# 원-핫 인코딩
## 반드시 2차원 배열을 사용
## 원-핫 인코딩은 보통 Feature 에 사용되기 때문
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

# 2차원 ndarray로 변환
items_arr = np.array(items).reshape(-1,1) #파이선 리스트를 넘파이 어레이 형식으로 변환
items_arr
```

```python
encoder = OneHotEncoder()

ohe_labels = encoder.fit_transform(items_arr) #fit_transform() : fit(데이터를 입력)하고나서 transform(데이터를 변환)하는
ohe_labels
```

```python
print(ohe_labels)
# 1의 위치를 압축된 표현으로 확인
```

```python
# 압축된 형식을 풀어서 표현
print(ohe_labels.toarray())
```

```python
# min-max scaling(정규화)
```

```python
import pandas as pd
movie_score = pd.DataFrame({
    "naver":[0, 2, 4, 6, 8],
    "netflix":[1, 2, 3, 4, 5]
})

movie_score
```

```python
from sklearn.preprocessing import MinMaxScaler

mm_scaler = MinMaxScaler()
movie_score_scaled = mm_scaler.fit_transform(movie_score)
movie_score_scaled # 최소값 0으로 최대값 1로 세팅
```

```python
# standard scaling(표준화)
# 평균이 0,표준편차가 1이 되도록 변환
```

```python
data1 = np.array([100, 110, 120, 130, 140])
data2 = np.array([0, 1, 2, 3, 4])
```

```python
data1.mean(), data1.var(), data1.std()
```

```python
data2.mean(), data2.var(), data2.std()
```

```python
from sklearn.preprocessing import StandardScaler

std_scaler1 = StandardScaler()
std_scaler2 = StandardScaler()
```

```python
data1_scaled = std_scaler1.fit_transform(data1.reshape(-1,1))
data2_scaled = std_scaler2.fit_transform(data2.reshape(-1,1))
```

```python
print(data1_scaled)
print()
print(data2_scaled)
```

```python
data1_scaled.mean(), data1_scaled.std()
```

```python
data2_scaled.mean(), data2_scaled.std() # 동일선상에 두고 데이터 비교 가능하도록
```
